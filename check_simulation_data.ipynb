```python
import numpy as np
import xarray as xr
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import geopandas as gpd
import sys
sys.path.append('/nfs/see-fs-02_users/earlacoa/wrf-analysis/misc/')
from cutshapefile import transform_from_latlon, rasterize
params = {
    'text.latex.preamble': ['\\usepackage{gensymb}'],
    'axes.grid': False,
    'savefig.dpi': 700,
    'font.size': 12,
    'text.usetex': True,
    'figure.figsize': [5, 5],
    'font.family': 'serif',
}
matplotlib.rcParams.update(params)
```

```python
path = '/nfs/b0122/Users/earlacoa/paper_aia_china'
sims = [f't{sim}' for sim in range(1, 56)]
months = [f'{month:02d}' for month in range(1, 13)]
regions = {'china': '/nfs/a68/earlacoa/shapefiles/china/china_taiwan_hongkong_macao.shp'}

pm25_values = {}

for region, shapefile in regions.items():
    shp = gpd.read_file(shapefile)
    shapes = [(shape, n) for n, shape in enumerate(shp.geometry)]
    
    for sim in sims:
        for month in months:
            with xr.open_dataset(f'{path}/{sim}/wrfout_d01_global_0.25deg_2015-{month}_PM2_5_DRY.nc') as ds:
                pm25 = ds['PM2_5_DRY']

            clip = rasterize(shapes, pm25.coords, longitude='lon', latitude='lat')
            clipped_ds = pm25.where(clip==0, other=np.nan)

            pm25_values.update({f'{sim}_{month}_{region}': clipped_ds.mean(dim='lat').mean(dim='lon')})
            
            fig = plt.figure(1, figsize=(8, 6))
            gs = gridspec.GridSpec(1, 1)
            pm25_values[f'{sim}_{month}_{region}'].plot()
            plt.title(f'Sim: {sim}, 2015-{month}, China-mean')
            plt.ylabel('Ambient PM$_{2.5}$ concentrations (${\mu}g$ $m^{-3}$)', fontsize=14)
            plt.xlabel('Date', fontsize=14)
            plt.savefig(f'/nfs/b0122/Users/earlacoa/png/paper_aia_emulator/checks/{sim}_{month}_{region}.png', dpi=700, alpha=True, bbox_inches='tight')
            plt.show()
```
